```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv("/content/tmdb_5000_movies.csv")
credits=pd.read_csv("/content/tmdb_5000_credits.csv")
```

Merge the movies and credits data


```python
movies = movies.merge(credits,on='title')
```

```python
movies= movies[['movie_id_y','title','overview','genres','keywords','cast_x','crew_x']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.iloc[0].genres
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L

```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert(obj):
  L=[]
  counter =0

  for i in ast.literal_eval(obj):
    if counter !=3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

```python
movies['cast']=movies['cast_x'].apply(convert)
```

```python
movies.head()
```

```python
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
       L.append(i['name'])
       break
  return L
```

```python
movies['crew_x']=movies['crew_x'].apply(convert)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew_x']=movies['crew_x']=movies['crew_x'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew_x']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id_y','title','tags']]
```
